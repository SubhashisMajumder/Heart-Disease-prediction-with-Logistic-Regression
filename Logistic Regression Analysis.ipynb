## Building Logistic Regression Model

### Train Test split

```python
dataset.drop('heart_diagnosis',axis=1).head()
```

```python
dataset['heart_diagnosis'].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(dataset.drop('heart_diagnosis',axis=1),dataset['heart_diagnosis'],test_size=0.20,random_state=50)
```

# Training & Predicting

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test,predictions)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,predictions)
```
